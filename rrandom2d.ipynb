```
import random
import numpy as np
import matplotlib.pyplot as plt
```

```
N = 30
```

```
d =1;
x = np.zeros(N+1)
y = np.zeros(N+1)
x[0]=0; y[0]=0
```

```
for i in range(0, N, 1):
  r = random.random()  # [0, 1)
  d = random.expovariate(i+5.7);
  if 0<= r <0.25: y[i+1]=y[i]+d; x[i+1]=x[i]
  elif 0.25<= r < 0.5: y[i+1]= y[i]; x[i+1]= x[i]+d
  elif 0.5<= r < 0.75 : y[i+1]= y[i]-d; x[i+1]= x[i]
  else: y[i+1]= y[i]; x[i+1]=x[i]-d
```

```
plt.plot(x, y, 'r--', x[0], y[0], 'bo', x[-1], y[-1], 'b*')
plt.xlabel('East->'); plt.ylabel('North->')
plt.rcParams["figure.figsize"] = (20,10)
plt.show()
```
